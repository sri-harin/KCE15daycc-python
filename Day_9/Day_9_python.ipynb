```python
#MIN BILL RS 50
units = int(input("Enter units consumed: "))

bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

# Minimum charge
if bill < 50:
    bill = 50

print("Total Electricity Bill:", bill)
```

```python
#15% CHARGE
units = int(input("Enter units consumed: "))

bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

# Minimum charge
if bill < 50:
    bill = 50

# Surcharge
if bill > 400:
    bill += bill * 0.15

print("Total Electricity Bill:", bill)
```

```python
#GREEN ENERGY
units = int(input("Enter units consumed: "))

bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

if units < 100:
    bill -= bill * 0.20

print("Total Electricity Bill:", bill)
```

```python
#FIXED LINE RENT
units = int(input("Enter units consumed: "))

bill = 0

# Base bill calculation
if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

# Green Energy Subsidy (20%)
if units < 100:
    bill -= bill * 0.20

# Minimum charge
if bill < 50:
    bill = 50

# Surcharge (15%)
if bill > 400:
    bill += bill * 0.15

# Fixed Line Rent
bill += 15

print("Total Electricity Bill:", bill)
```

```python
#A GRAND BILL SUMMARY

units = int(input("Enter units consumed: "))

# Base bill calculation
if units <= 100:
    base = units * 1.5
elif units <= 200:
    base = 100 * 1.5 + (units - 100) * 2.5
else:
    base = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4

bill = base

# Green Energy Subsidy (20%)
if units < 100:
    bill -= bill * 0.20

# Minimum charge
if bill < 50:
    bill = 50

# Surcharge (15%)
surcharge = 0
if bill > 400:
    surcharge = bill * 0.15
    bill += surcharge

# Fixed Line Rent
line_rent = 15
bill += line_rent

# Tax (example: 5%)
tax = bill * 0.05

total = bill + tax

# Grand Bill Summary
print(f"Base: {base} | Surcharge: {surcharge} | Tax: {tax} | Total: {total}")
```
