```python
#SI
P = float(input("Principal: "))
R = float(input("Rate: "))
T = float(input("Time: "))

SI = (P * R * T) / 100
print("Simple Interest:", SI)
```

```python
#TOTAL AMOUNT
Amount = P + SI
print("Total Amount:", Amount)
```

```python
#MONTHLY INSTALLMENT
loan = float(input("Loan Amount: "))
years = int(input("Years: "))

months = years * 12
monthly_payment = loan / months

print("Monthly Payment:", monthly_payment)
```

```python
#COMPOUND INTEREST
import math

P = float(input("Principal: "))
R = float(input("Rate: "))
T = float(input("Time: "))

A = P * math.pow(1 + R / 100, T)
print("Compound Amount:", A)
```

```python
#DIFFREENCE BETWEEN CI AND SI
SI = (P * R * T) / 100
CI = A - P

print("Difference (CI - SI):", CI - SI)
```

```python
#THE DOUBLING RULE
import math

P = float(input("Principal: "))
A = P * math.pow(1 + 7 / 100, 10)

print("Amount after 10 years:", A)
```

```python
#DAILY INTEREST CALCULATION
P = float(input("Principal: "))
annual_rate = float(input("Annual Rate: "))
days = int(input("Number of days: "))

daily_rate = annual_rate / 365
interest = (P * daily_rate * days) / 100

print("Interest for given days:", interest)
```
